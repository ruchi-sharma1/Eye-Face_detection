```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
face= cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
eye= cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_eye.xml')
smile= cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_smile.xml')
```

```python
img= cv2.imread('/content/face3.jpg')
```

```python
gray= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY )
faces = face.detectMultiScale(gray, 1.3, 5)
```

```python
for (x,y,w,h) in faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(200,155,190),5)
  cv2.putText(img,'face',(x, y-5),cv2.FONT_HERSHEY_SIMPLEX,2,(200,155,190),7) 
  gray = gray[y:y+h, x:x+w]
  roi_color = img[y:y+h, x:x+w]
  eyes = eye.detectMultiScale(gray,1.2,5)
  for (ex,ey,ew,eh) in eyes:
     cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,250,0),2)
     cv2.putText(roi_color,'eye', (ex,ey-5), cv2.FONT_HERSHEY_SIMPLEX,1, (0,250,0),2)

  smiles= smile.detectMultiScale(gray,1.5,7)
  for (sx,sy,sw,sh) in smiles:
    cv2.rectangle(roi_color,(sx,sy),(sx+sw,sy+sh),(0,0,100),2)
    cv2.putText(roi_color,'smile', (sx-5,sy), cv2.FONT_HERSHEY_SIMPLEX,1, (0,0,100),2)
```

```python
cv2_imshow(img)
```
